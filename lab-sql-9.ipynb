```python
# 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
```

```python
query2 = 'select *, dayname(rental_date) as rentalday, monthname(rental_date) as rentalmonth, \
dayname(return_date), monthname(return_date), \
case \
when dayname(rental_date) in ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday") then "workday" \
else "weekend" \
end as day_type \
from rental'
```

```python
# 6. Which kind of movies (rating) have a mean duration of more than two hours?
```

```python
query6 = "select c.name as 'Category Name', (f.length/60)  as hour from film as f \
join film_category as fc \
on f.film_id = fc.film_id \
join category as c \
on fc.category_id = c.category_id \
group by c.name \
having hour > 2;"
```

```python
result6 = pd.read_sql_query(query6, engine)
```

```python
display(result6)
```

```python
# 7. Rank films by length (filter out the rows that have nulls or 0s in length column).
# In your output, only select the columns title, length, and the rank.
```

```python
query7 = "select \
	title, \
	length, \
    dense_rank() over(order by length) as ranking \
 from film where length != 0 and length is not null"
```
